```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Matplotlib is a basic plotting library that gives full control but needs more code.

Seaborn is built on Matplotlib and makes beautiful charts easily with less code.

```python
df =pd.read_csv("/content/Datasets/Titanic_Dataset/Titanic-Dataset.csv")
```

```python
df['Age']= df['Age'].fillna(df['Age'].median())
```

```python
df.isnull().sum()
```

```python
df['has_Cabin']=df['Cabin'].notna().astype(int)
df.drop('Cabin',axis=1,inplace=True)
```

```python
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
sns.countplot(x='Survived',data=df,palette='dark')
plt.title('Survived vs not Survived')
plt.show()
```

```python
sns.countplot(x='Pclass', data=df, palette='rainbow')
plt.title('Passenger Class')
plt.show()
```

```python
sns.countplot(x='Sex', data=df, palette='Set1')
plt.title('Sex')
plt.show()
```

```python
sns.countplot(x='SibSp', data=df, palette='Set1')
plt.title('SibSp')
plt.show()
```

```python
sns.countplot(x='Parch', data=df, palette='Set1')
plt.title('Parch')
plt.show()
```

```python
sns.countplot(x='has_Cabin',data=df, palette='Set1')
plt.title('has_Cabin')
plt.show()
```

```python
sns.histplot(x='Fare',data=df,color='b')
plt.title
```

```python
sns.histplot(x='Age',data=df,color='b')
plt.title('Age')
plt.show()
```

```python
sns.barplot(x='Pclass',y='Fare',data=df,palette='rainbow')
plt.title('Pclass vs Fare')
plt.show()
```

```python
n_df=df
```

```python
n_df
```

```python
n_df.drop('Name',axis=1,inplace=True)
```

```python
n_df.drop('Ticket',axis=1,inplace=True)
```

```python

Mapping_dict={'male':0,'female':1}
n_df['Sex']=n_df['Sex'].map(Mapping_dict)
```

```python
mapping={'S':0,'C':1,'Q':2}
n_df['Embarked']=n_df['Embarked'].map(mapping)
```

```python
n_df
```

```python
n_df.info()
```

```python
n_df.set_index('PassengerId', inplace=True)
```

```python
n_df
```

```python
numeric_cols = n_df.select_dtypes(include=np.number)
correlation_matrix = numeric_cols.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=n_df)
```

```python
sns.catplot(x='Pclass', y='Survived', hue='Sex', data=df,kind='bar')
plt.title('Survival Rate by Pclass and Sex')
plt.show()
```

```python
sns.boxplot(x='Pclass', y='Age', data=n_df)
plt.title('Age Distribution by Passenger Class')
plt.show
```
